# DataBricks CLI
- he Databricks command-line interface (CLI) provides an easy-to-use interface to the Databricks platform.
**For example, you can use the Databricks CLI to do things such as:**
- Provision compute resources in Databricks workspaces.
- Run data processing and data analysis tasks.
- List, import, and export notebooks and folders in workspaces.

## setup the CLI
- **Requirements:**
    - python 3.6 and above.

### Install CLI
```python
pip install databricks-cli
```

### Upgrade CLI
```python
pip install databricks-cli --upgrade
```

### Check Version
```cmd
databricks --version

or 

databricks -v
```



# Authentication Of CLI with Databricks.
- Before you can run CLI commands, you must set up authentication. To authenticate to the CLI you use a Databricks personal access token.

#### Get Personal Access Token Of DataBricks

```cmd
databricks configure --token
```
The command begins by issuing the prompt:

```cmd
Databricks Host (should begin with https://): **Mention your databricks host name**
Token: **Mention Token**
```
